```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
%matplotlib inline
```

```python
def unpickle(file):
    import pickle
    with open(file, 'rb') as fo:
        dict = pickle.load(fo, encoding='bytes')
    return dict
```

```python
meta = unpickle('/Users/dmitrijvolkov/Neural_Networks_from_Scratch/cifar-10-batches-py/batches.meta')
data = unpickle('/Users/dmitrijvolkov/Neural_Networks_from_Scratch/cifar-10-batches-py/data_batch_1')
pics = data[b'data']
labels = data[b'labels']
meta = meta[b'label_names']
```

```python
def img_plot(pics, ind):
    img = pics[ind]
    print(meta[labels[ind]].decode())
    plt.imshow(img.reshape(3,32,32).transpose(1,2,0), interpolation='lanczos')
    plt.show()
```

```python
img_plot(pics, 100)
```
